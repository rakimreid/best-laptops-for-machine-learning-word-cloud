```python
# import all necessary modules

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pathlib import Path
```

```python
# Import text file

df = pd.read_excel("C:/Portfolio/Data Engineer Track/Python/Best Laptop For Machine Learning Text Analytics/article.xlsx")

```

```python
# According to this article: It seems the best laptops for machine learning have - NVIDIA GeForce graphics especially GeForce GTX; Intel processors with Core i7 processors; and there is a significant focus on the amount of RAM. 
# and focus on power, the display, performance and over battery life and other quality of life concerns.
```

```python
# First attempt with the raw text:
text = df.description[0]

# Create and generate a word cloud image:
wordcloud = WordCloud(background_color = "white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["laptop", "machine", "learning", "AI", "programming", "buy", "Amazon", "list", "offer", "great", "along", "offers", "comes", "come", "need", "much", "besides", "Besides", "range", "next", "make", "makes", "thing", "will", "consider", "getting", "another"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
